```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## Visualization
```

```python
# histogram shows frequency of a numerical variable.
plt.figure(figsize=(6, 4))
sns.histplot(df['Area'])
plt.title('Distribution of Area')
plt.ylabel('Frequency')
plt.xlabel('Area')

plt.show()
```

```python
#KDE for density distribution of a numerical variable
sns.kdeplot(df['Area'])
plt.show()
```

```python
# box plot for spread of data and outliers.
sns.boxplot(x=df['Area'])

plt.tight_layout()
plt.show()
```

```python
# violinplot Shows the distribution of a numerical variable across categories.
sns.violinplot(x='Class', y='Area', data=df)

plt.tight_layout()
plt.show()
```

```python
#countplot Shows the count of each category
sns.countplot(x='Class', data=df)

plt.show()
```

```python
#barplot Shows the average of a numerical variable for each category
sns.barplot(x='Class', y='Area', data=df)

plt.show()
```

```python
# scatterplot shows the relationship between two numerical variables.
plt.figure(figsize=(6, 4))

sns.scatterplot(x='Area',
                y='Perimeter',
                hue='Class',
                data=df)

plt.title('Area vs Perimeter by Bean Class')
plt.xlabel('Area')
plt.ylabel('Perimeter')

plt.legend(title='Bean Type')

plt.show()
```

```python
# heat map visualizes the correlation matrix.
plt.figure(figsize=(6, 4))
sns.heatmap(df.corr(numeric_only=True))
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Jointplot shows the relationship between two variables along with their distributions
sns.jointplot(x='Area', y='Perimeter', data=df)
plt.show()
```
